```python
import matplotlib.pyplot as plt
import numpy             as np
import seaborn           as sns

sns.set_style('darkgrid')
sns.mpl.rc('figure',figsize=(16, 6))
```

```python
x = np.arange(0,9)

bsolar_cd_time = np.round(np.array([1/2.03, 1/1.74, 1.14, 2.26, 3.63, 5.45, 9.21, 11.35, 21.23]), 2)

print(bsolar_cd_time)
```

```python
bsolar_la_time = np.round(np.array([1/2.75, 1/2.65, 1/1.63, 1.12, 1.99, 3.23, 5.22, 7.23, 13.98]), 2)

print(bsolar_la_time)
```

```python
bolasso_cd_time = np.round(np.array([13.84, 18.14, 46.10, 130.67, 265.29, 459.31, 939.53, 1752.76, 2927.86]), 2)

print(bolasso_cd_time)
```

```python
fig = plt.figure()

x_axi = [r'$100 \times 100$', r'$200 \times 200$', r'$400 \times 400$', r'$600 \times 600$', r'$800 \times 800$', r'$1000 \times 1000$', r'$1200 \times 1200$', r'$1400 \times 1400$', r'$1600 \times 1600$']

plt.plot(x, bsolar_cd_time , linewidth=5, markersize=12, color='b', ls ='--', marker='o', label='bsolar-3')
plt.plot(x, bolasso_cd_time, linewidth=5, markersize=12, color='r', ls ='--', marker='o', label='bolasso')

plt.legend(loc='best', shadow=True, prop={'size': 22})

plt.xlabel('matrix size', fontsize=22)
plt.ylabel('time (second)', fontsize=24)

plt.xticks(x, x_axi, fontsize=22, rotation=45)
plt.yticks(fontsize=22)

plt.xlim(0, 8)
plt.tight_layout()

plt.show()

fig.savefig("runtime.pdf", bbox_inches='tight')
```
